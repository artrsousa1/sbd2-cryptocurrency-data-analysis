```python
total = len(df)
print(f'Total: {total}')

duplicated = df.duplicated().sum()
print(f'Duplicados: {duplicated}')

print(f'Únicos: {total - duplicated}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x, y, hue = 'porcentagem', 'total', 'dados'
hue_order = ['únicos', 'duplicados']
data = {
  'dados': ['únicos', 'duplicados'],
  'total': [total - duplicated, duplicated],
}
data['porcentagem'] = [v / total * 100 for v in data['total']]

sns.barplot(data=data, x=x, y=y, hue=hue, hue_order=hue_order)
plt.ylabel('Total de dados')
plt.xlabel('Tipo de dado')
plt.title('Diferença percentual entre valores únicos e duplicados')
plt.ylim(0, 100)
plt.yticks([0, 100], ['0%', '100%'])
plt.show()
```
